```
from pyspark.context import SparkContext
```

```
sc=SparkContext.getOrCreate()
```

```
#loading csv file
txt=sc.textFile("/FileStore/tables/Stores.csv")
```

```
#Action Methods
#Displays total numbers of elements in the Stores.csv file
txt.count()
```

```
#retrieve all elements of the dataset
txt.collect()
```

```
#Displays first 3 row from Stores.csv
txt.take(3)
```

```
#Display top 1 row of the dataset
txt.top(1)
```

```
txt.cache()
```

```
#Displays minimum values from each row
txt.min()
```

```
#Transformation Methods
# Map a function/lambda expression to each line using split method
# collect the results.
txt.map(lambda line: line.split()).collect()
```

```
txt.max()
```

```
# Display all data as a single flat map by using collect()
txt.flatMap(lambda line: line.split()).collect()
```

```
#Filter out the results having cash word and displays the result
search=txt.filter(lambda line:"Store_Area" in line)
search.collect()
```

```
#To drop/remove the duplicate rows in dataset we use distinct function
txt.distinct().collect()
```

```
#Creating RDD using parallelize
rdd=sc.parallelize(['All','Project','operations','performed','above'])
rdd.collect()
```

```
#Actions Performed
print("Number of Partitions: "+str(rdd.getNumPartitions()))
print("Action: Total number of elements: "+str(rdd.count()))
```

```
rdd.collect()
```

```
#Displays the maximum element 
rdd.max()
```
